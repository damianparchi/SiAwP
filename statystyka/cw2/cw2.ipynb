```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli, binom, poisson, norm, kurtosis, skew

# Dane
values = np.array([1, 2, 3, 4, 5, 6])
probabilities = np.array([1/6, 1/6, 1/6, 1/6, 1/6, 1/6])
n = 100

# Rozkład Bernoulliego
bernoulli_samples = bernoulli.rvs(p=probabilities[0], size=n)

# Rozkład Dwumianowy
binomial_samples = binom.rvs(n=6, p=probabilities[0], size=n)

# Rozkład Poissona
poisson_samples = poisson.rvs(mu=values.mean(), size=n)

# Statystyki podstawowe
def calculate_basic_stats(samples):
    mean = np.mean(samples)
    variance = np.var(samples)
    kurt = kurtosis(samples)
    skewness = skew(samples)
    return mean, variance, kurt, skewness

# Oblicz statystyki dla każdego rozkładu
stats_bernoulli = calculate_basic_stats(bernoulli_samples)
stats_binomial = calculate_basic_stats(binomial_samples)
stats_poisson = calculate_basic_stats(poisson_samples)

print("Statystyki podstawowe dla rozkładu Bernoulliego:", stats_bernoulli)
print("Statystyki podstawowe dla rozkładu Dwumianowego:", stats_binomial)
print("Statystyki podstawowe dla rozkładu Poissona:", stats_poisson)
```

```python
# Wykresy rozkładów prawdopodobieństwa
plt.figure(figsize=(15, 5))

# Rozkład Bernoulliego
plt.subplot(1, 3, 1)
plt.bar(values, bernoulli.pmf(values, probabilities[0]), color='blue', alpha=0.7, label='Bernoulli PMF')
plt.title('Rozkład Bernoulliego')
plt.legend()

# Rozkład Dwumianowy
plt.subplot(1, 3, 2)
plt.bar(values, binom.pmf(values, 6, probabilities[0]), color='green', alpha=0.7, label='Binomial PMF')
plt.title('Rozkład Dwumianowy')
plt.legend()

# Rozkład Poissona
plt.subplot(1, 3, 3)
plt.bar(values, poisson.pmf(values, values.mean()), color='red', alpha=0.7, label='Poisson PMF')
plt.title('Rozkład Poissona')
plt.legend()

plt.show()
```

```python
# Generowanie danych dla rozkładu normalnego
normal_samples = norm.rvs(loc=0, scale=2, size=100)

# Statystyki podstawowe
stats_normal = calculate_basic_stats(normal_samples)

print("Statystyki podstawowe dla rozkładu normalnego:", stats_normal)
```

```python
# Generowanie większej liczby danych dla rozkładu normalnego
larger_normal_samples = norm.rvs(loc=0, scale=2, size=1000)

# Statystyki podstawowe
stats_larger_normal = calculate_basic_stats(larger_normal_samples)

print("Statystyki podstawowe dla rozkładu normalnego (większa liczba danych):", stats_larger_normal)
```

```python
# Histogram dla rozkładu normalnego
plt.hist(normal_samples, bins=20, density=True, alpha=0.5, color='blue', label='Normal (mean=0, std=2)')
plt.hist(larger_normal_samples, bins=20, density=True, alpha=0.5, color='green', label='Normal (mean=0, std=2) - Larger Sample')
plt.plot(np.linspace(-6, 6, 100), norm.pdf(np.linspace(-6, 6, 100), loc=0, scale=2), 'r-', label='Normal PDF (mean=0, std=2)')
plt.plot(np.linspace(-6, 6, 100), norm.pdf(np.linspace(-6, 6, 100), loc=-1, scale=0.5), 'k--', label='Normal PDF (mean=-1, std=0.5)')
plt.title('Histogram i Wykresy dla Rozkładu Normalnego')
plt.legend()
plt.show()
```
